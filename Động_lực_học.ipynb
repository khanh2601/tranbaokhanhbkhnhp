```
def pose(theta, alpha, a, d):

  r11 = cos(theta)
  r12 = -sin(theta) * cos(alpha)
  r13 = sin(theta) * sin(alpha)
  r21 = sin(theta)
  r22 = cos(theta) * cos(alpha)
  r23 = - sin(alpha) * cos(theta)
  r32 = sin(alpha)
  r33 = cos(alpha) 
  x = a * cos(theta)
  y = a * sin(theta)
  z = d 

  T = Matrix([
    [r11, r12, r13, x],
    [r21, r22, r23, y],
    [0.0, r32, r33, z],
    [0.0, 0.0, 0.0, 1]
  ])
  
  T = simplify(T)

  return T
```
